```python
#Slope One和KNN-Slope One的MAE
#导入库
import numpy as np
import pylab as pl
#设置数据
x1 = [40,60,80,100,120,140,160,180,200]
y1 = [0.75, 0.74, 0.738, 0.737, 0.737,0.736, 0.737, 0.738,0.74]
x2 = [40,60, 80, 100, 120, 140,160,180,200]
y2 = [0.74, 0.74,0.741, 0.74,0.74,0.74,0.739,0.74,0.74]
#设置横纵坐标的刻度
pl.xlim(40, 200)# set axis limits
pl.ylim(0.73, 0.76)
#设置横纵坐标的显示
pl.xlabel('k')# make axis labels
pl.ylabel('MAE')
#设置图中线的显示
pl.plot(x1, y1, 'rv-',label='KNN Slope One')# use pylab to plot x and y
pl.plot(x2, y2, 'gs-',label='Slope One')
#设置显示数据点的含义 
pl.legend(loc="upper right")
#保存图片
pl.savefig("1.jpg")
#展示图片 
pl.show()
```

```python
#各种算法的MAE

#导入库
import numpy as np
import pylab as pl

#设置数据
num_list = [0.815,0.818,0.75,0.73]
#设置y轴数据范围
pl.ylim(0.7, 0.83)
#设置x轴显示
name_list = ['UBCF','IBCF','EMDP','KNN & Slope One']
#画图中数据和柱状图的颜色
pl.bar(range(len(num_list)), num_list,color='rgb',tick_label=name_list) 
#设置横纵轴的显示
pl.xlabel("Algorithm")  
pl.ylabel("MAE")
#保存图片
pl.savefig("2.jpg") 
#显示图片
pl.show()
```
