```python
#KNN和KNN改进的MAE
#导入库
import numpy as np
import pylab as pl
#设置数据
x1 = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
y1 = [0.68,0.68,0.68,0.69,0.71,0.75,0.80,0.83,0.92]
x2 = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
y2 = [0.72,0.73,0.74,0.75,0.77,0.80,0.85,0.87,0.95]
#设置横纵坐标的刻度
pl.xlim(0, 1.0)# set axis limits
pl.ylim(0.6,1.0)
#设置横纵坐标的显示
pl.xlabel('a')# make axis labels
pl.ylabel('MAE')
#设置图中线的显示
pl.plot(x1, y1, 'rv-',label='IF-KNN')# use pylab to plot x and y
pl.plot(x2, y2, 'gs-',label='KNN ')
#设置显示数据点的含义 
pl.legend(loc="upper left")
#保存图片
pl.savefig("KNNCF_MAE.jpg")
#展示图片 
pl.show()
```

```python
#Slope One和KNN-Slope One的MAE
#导入库
import numpy as np
import pylab as pl
#设置数据
x1 = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
y1 = [0.70,0.70,0.71,0.72,0.73,0.76,0.79,0.84,0.93]
x2 = [0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
y2 = [0.75,0.76,0.77,0.79,0.83,0.87,0.90,0.93,0.97]
#设置横纵坐标的刻度
pl.xlim(0, 1.0)# set axis limits
pl.ylim(0.6,1.0)
#设置横纵坐标的显示
pl.xlabel('a')# make axis labels
pl.ylabel('MAE')
#设置图中线的显示
pl.plot(x1, y1, 'rv-',label='KNN-WSO')# use pylab to plot x and y
pl.plot(x2, y2, 'gs-',label='WSO')
#设置显示数据点的含义 
pl.legend(loc="upper left")
#保存图片
pl.savefig("SOCF_MAE.jpg")
#展示图片 
pl.show()
```

```python
#各种算法的RMSE

#导入库
import numpy as np
import pylab as pl

#设置数据
num_list = [1.3537,1.3392,1.3164,1.3046,1.2686]
#设置y轴数据范围
pl.ylim(1.0, 1.4)
#设置x轴显示
name_list = ['UBCF','IBCF','EMDP','KNNCF','SOCF']
#画图中数据和柱状图的颜色
pl.bar(range(len(num_list)), num_list,color='rgb',tick_label=name_list) 

#设置横纵轴的显示
pl.xlabel("Algorithm")  
pl.ylabel("RMSE")
#保存图片
pl.savefig("all_RMSE.jpg") 
#显示图片
pl.show()
```
